## 1. Minimal ADK Project Structure

From the [Quickstart Guide](https://google.github.io/adk-docs/get-started/quickstart/), here's the simplest setup:

```
my_agent/
├── __init__.py     # make it a Python package and import agent
├── agent.py        # Your agent definition
└── .env           # For API keys (GOOGLE_API_KEY=your_key)
```

### Key Files:
- **`agent.py`**: Contains your agent code (shown next).
- **`.env`**: Stores sensitive data like API keys.

### Example `agent.py`:
```python
from google.adk.agents import Agent

weather_agent = Agent(
    name='weather_agent',
    model='gemini-1.5-flash',
    description='Provides weather updates'
)
```

## 2. Running Your Agent

### Header Bar

-   **Session ID** (e.g., 9401d62a-caff-492b-b6e9-d19001d891be):

    -   Tracks your current chat session. It's unique for each conversation, helping you manage multiple tests or debug sessions.

-   **Token Streaming Toggle** (labeled 'TOKEN STREAMING', gray/off in your image):

    -   When on, responses stream character-by-character, showing the agent's thought process live. When off, you get the full response at once.

-   **New Session Button** (labeled '+ NEW SESSION'):

    -   Starts a fresh conversation, keeping your current session intact for reference.

### Left Panel: Navigation & Evaluation

-   **Tabs** (Events, State, Artifacts, Sessions, Eval):

    -   **Events**: Logs all agent actions, like tool calls or model queries, for debugging.

    -   **State**: Shows the agent's current memory or context, like stored variables.

    -   **Artifacts**: Lists generated files or data, like reports or images.

    -   **Sessions**: Displays all active/past sessions for easy switching.

    -   **Eval**: Tests your agent's performance with evaluation sets.

-   **Eval Tab Details** (active in your screenshot):

    -   **All Eval Sets**: Shows '0' in your image---no sets created yet.

    -   **Create New Evaluation Set** (blue button): Lets you define test cases (e.g., queries and expected outputs).

    -   **Description**: Explains evaluation sets as curated collections for testing agent accuracy.

    -   **Second Create Evaluation Set Button**: A duplicate for convenience at the bottom.

### Right Panel: Chat Interface

-   **Chat Area**:

    -   Shows your queries and responses. In your screenshot:

        -   Query 1: 'What's the weather like in NYC?' failed (get_weather with lightning bolt), with an apology: 'I am sorry, I cannot fulfill this request...'.

        -   Query 2: 'Can you tell me the weather in new york?' succeeded (get_weather with checkmark), responding: 'OK. The weather in New York is sunny with a temperature of 25 degrees Celsius (77 degrees Fahrenheit).'

    -   Displays intermediate steps, like tool calls, for transparency.

-   **Input Field** (labeled 'Type a Message...'):

    -   Supports text queries and multimodal inputs via:

        -   **Paperclip Icon**: Upload files for processing.

        -   **Mic Icon**: Likely for voice input (browser-dependent).

        -   **Camera Icon**: Upload images for analysis, showing ADK's multimodal capabilities.
